There are a few diffrent wasy to plot in julia(including calling PyPlot)

Here we'll show you how to use Plots.jl

```julia
using Pkg
Pkg.add("Plots") 
Pkg.add("PlotlyJS")# intalacion de la biblioteca
```

```julia
using Plots # uso de la biblioteca
```

One of the advantages to Plots.jl is that is allows you to semlessly change backends. In this notebook, we'll try out the gr() and plotlyjs() backends.

first, let's generate some artificial data to plot!

```julia
x = -3: 0.1 : 3
f(x) = x ^ 2
y = f.(x)
```

**Now let's load the GR backend**

```julia
gr()
```

```julia
plot(x, y, label = "line")
scatter!(x, y, label="points")
```

Ahora con plotlyjs()

```julia
plotlyjs()
```

```julia
plot(x, y, label="line")
scatter!(x, y, label="points")
```

```julia
xlabel!("Number of Pirates [Approximate]")
ylabel!("Global Temperature (C)")
title!("Influence of pirate population on global warming")
     
```
